Przykład 4

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("pacjent.txt", delim_whitespace=True, header=0)
print(df.head())
```

```python
# wariant I
fig = px.scatter_matrix(df,dimensions=df.columns[0:4],color='group')
fig.show()
```

```python
# wariant II
sns.pairplot(df, hue='group', palette='dark')
plt.show()
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import pandas as pd

df = pd.read_csv("dent.txt", delim_whitespace=True, header=0)
print(df.head())
```

```python
fig = px.parallel_coordinates(df, dimensions=df.columns[0:12], color='Grupa')
fig.show()
```

```python
sns.heatmap(df.corr())# korelacja - mapa cieplna
plt.show()
```

```python
fig = px.scatter_matrix(df,dimensions=df.columns[0:12],color='Grupa')
fig.show()
```

```python
sns.pairplot(df, hue='Grupa', palette='dark')
plt.show()
```
