Przykład 1

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.50,random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50, c='grey')
plt.show()

kmeans = KMeans(n_clusters=3, n_init=10, init = 'random') #k-means++
kmeans.fit(X) # uczenie modelu
y_kmeans = kmeans.predict(X) # predykcja numerów skupień

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_# obliczone przez kmeans punkty centralne skupień

plt.scatter(centers[:, 0], centers[:, 1], c='black', s=80, alpha=0.5)
plt.show()

print('SSE: ', kmeans.inertia_) # poszukujemy jak najmniejszej wartości

#0.2 - SSE 23
#0.3 - SSE 53
#0.5 - SSE 147.47
#0.7 - SSE 286.34

#SSE rosnie gdy inne parametry rosną
```

```python
print(y_true)
print(y_kmeans)
```

Przykład 3

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

X, y = make_moons(200, noise=.05, random_state=0)
y_kmeans = KMeans(2, random_state=0).fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.show()

print('y_true: ')
print(y) # numery skupień dla każdego obiektu (dane wejściowe)
print('y_kmeans: ')
print(y_kmeans) # numery skupień dla każdego obiektu według kmeans
print('Accuracy k-means: ', accuracy_score(y,y_kmeans))
```

```python
model = DBSCAN(eps=0.2, min_samples=5, metric ='cityblock')# odpowiednie dobranie eps pozawala na lepsze grupowanie, dystans miedzy punktami zeby byly brane za sasiadow
y_dbscan = model.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, s=50, cmap='viridis')
plt.show()

print('y_dbscan: ')
print(y_dbscan) # numery skupień dla każdego obiektu według dbscan
print('Accuracy DBSCAN: ', accuracy_score(y,y_dbscan))
```

Przykład 4

```python
import plotly.express as px
import numpy as np

df = px.data.iris()
X = df.values[:,0:4]
y = df.values[:,5]

print(df.head(5))

y_kmeans = KMeans(3, n_init=5, init = 'k-means++', random_state=0).fit_predict(X)

#y_kmeans = DBSCAN(eps=1, min_samples=2, metric ='manhattan').fit_predict(X)

y_kmeans[np.where(y_kmeans==2)]=3
y_kmeans[np.where(y_kmeans==0)]=2

print(y_kmeans)
print(y)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
plt.show()

accuracy_score(np.int64(y), y_kmeans)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import seaborn as sns

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data')

X = np.append(df.values[:,0:1], df.values[:,2:], axis=-1)

le = LabelEncoder()
y = le.fit_transform(df.values[:,1])

y_kmeans = KMeans(2, n_init=5, init = 'k-means++', random_state=0).fit_predict(X)

print(y_kmeans)
print(y)

accuracy_score(np.int64(y), y_kmeans)
```
