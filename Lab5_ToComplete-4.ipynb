```python
class IrisFlower:
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width):
        self.sepal_length = sepal_length
        self.sepal_width = sepal_width
        self.petal_length = petal_length
        self.petal_width = petal_width

    def __repr__(self):
        return f"Iris: {self.sepal_length}, {self.sepal_width}, {self.petal_length}, {self.petal_width}"

    def area(self):
        return (self.sepal_length * self.sepal_width) + (self.petal_length * self.petal_width)
    
    def __lt__(self, other):
        return self.area() < other.area()
```

```python
from sklearn import datasets
```

```python
dataset = datasets.load_iris()
```

```python
print(dataset)
```

```python
data = dataset['data']
targets = dataset['target']
targets_names = dataset['target_names']
feature_names = dataset['feature_names']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, targets, test_size=0.33, random_state=42)

print(len(X_train), len(X_test))
print(y_train)
```

```python
df_train = pd.DataFrame(data=np.concatenate([X_train, y_train.reshape(-1, 1)], axis=1), columns=feature_names + ['class'])
df_train.head()
```

```python
sepal_length_class_1 = df_train[df_train['class'] == 0]['sepal length (cm)']
sepal_width_class_1 = df_train[df_train['class'] == 0]['sepal width (cm)']


sepal_length_class_2 = df_train[df_train['class'] == 1]['sepal length (cm)']
sepal_width_class_2 = df_train[df_train['class'] == 1]['sepal width (cm)']


sepal_length_class_3 = df_train[df_train['class'] == 2]['sepal length (cm)']
sepal_width_class_3 = df_train[df_train['class'] == 2]['sepal width (cm)']


petal_length_class_1 = df_train[df_train['class'] == 0]['petal length (cm)']
petal_width_class_1 = df_train[df_train['class'] == 0]['petal width (cm)']


petal_length_class_2 = df_train[df_train['class'] == 1]['petal length (cm)']
petal_width_class_2 = df_train[df_train['class'] == 1]['petal width (cm)']


petal_length_class_3 = df_train[df_train['class'] == 2]['petal length (cm)']
petal_width_class_3 = df_train[df_train['class'] == 2]['petal width (cm)']


fig, axes = plt.subplots(2, 1)
axes[0].scatter(sepal_length_class_1, sepal_width_class_1)
axes[0].scatter(sepal_length_class_2, sepal_width_class_2)
axes[0].scatter(sepal_length_class_3, sepal_width_class_3)
axes[1].scatter(petal_length_class_1, petal_width_class_1)
axes[1].scatter(petal_length_class_2, petal_width_class_2)
axes[1].scatter(petal_length_class_3, petal_width_class_3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

clf = LogisticRegression(random_state=0)
clf.fit(X_train[:, :2], y_train)
y_pred = clf.predict(X_test[:, :2])

print(classification_report(y_test, y_pred))
```

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components=3)
pca.fit(X_train)
X_train_transformed = pca.transform(X_train)

clf = LogisticRegression(random_state=0)
clf.fit(X_train_transformed, y_train)

pca.transform(X_test)
y_pred = clf.predict(pca.transform(X_test))

print(classification_report(y_test, y_pred))
```

```python
from sklearn import decomposition
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=60)
clf.fit(X_train, y_train)

pca.transform(X_test)
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn import decomposition
from sklearn.neighbors import KNeighborsClassifier

pca = decomposition.PCA(n_components=1)
pca.fit(X_train)
X_train_transformed = pca.transform(X_train)

clf = KNeighborsClassifier(n_neighbors=60)
clf.fit(X_train_transformed, y_train)

pca.transform(X_test)
y_pred = clf.predict(pca.transform(X_test))

print(classification_report(y_test, y_pred))
```
