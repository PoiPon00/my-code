Na koniec przykład wykorzystania pochodnych cząstkowych, w tym wypadku do wyznaczenia gradientu obrazu.

Żeby nie było za łatwo, tym razem policzymy gradient z wykorzystaniem masek Sobela (szczegóły na Technikach Obrazowania Medycznego). Dla naszych potrzeb zakładamy, że poprostu policzyliśmy gradient w kierunku x i y. Po laboratoriach, będą Państwo mogli podmienić Sobela na własne implementacje np. metodą różnic skończonych

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('agh.jpg',0)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
```

Wyświetlmy obraz

```python
fig = plt.figure(figsize= (20,20))
plt.subplot(1,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Obraz'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

Gradient jest wektorem, także możemy wyliczyc moduł i kąt

```python
mag=np.sqrt(sobelx*sobelx+sobely*sobely)
angle=np.arctan(sobelx,sobely)
```

```python
fig = plt.figure(figsize= (15,15))
plt.title('Moduł'), plt.xticks([]), plt.yticks([])
plt.imshow(mag,cmap = 'gray')
```
