```python
from scipy.cluster.hierarchy import dendrogram, linkage
import seaborn as sns
from seaborn import clustermap
import pandas as pd
from matplotlib import pyplot as plt

# wczytanie i podgląd danych
df = pd.read_csv("/content/drive/My Drive/Dane/pacjent.txt", delim_whitespace=True, header=0)
print(df.head())

#analiza skupień
df = df.set_index('group') # ustawienie etykiet obiektów
Z = linkage(df.values, method='ward', metric='euclidean')
# oblicza macierz odległości i ustala hierarchiczną
# relację pomiędzy obiektami

d = dendrogram(Z, labels=df.index, color_threshold=10)
# rysuje dendrogram (oś x - opis obiektów, oś y - odległość)

plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()

# dwuwymiarowa analiza skupień i mapa cieplna
clustermap(df, metric="euclidean", method="ward", cmap="Blues", annot = True, z_score = 1)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv("/content/drive/My Drive/Dane/dent.txt", delim_whitespace=True, header=0)
print(df.head())

df = df.set_index('Producent')
del df['Grupa'] # usuwanie kolumny – dlaczego?

Z = linkage(df.values, method='ward', metric='euclidean')
d = dendrogram(Z, labels=df.index, leaf_rotation=0, color_threshold=2000, above_threshold_color='grey') # uzupełnić

plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()

sns.clustermap(df, metric="correlation", method="single", cmap="Greens", z_score=0) # uzupełnić
plt.show()
```
