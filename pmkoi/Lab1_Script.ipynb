```python
#ZADANIE 3
#Zaproponuj metodę umożliwiającą zbliżone wyznaczenie wartości epsilon, tj. najmniejszej wartości, 
#która po dodaniu do jedności nie będzie skutkować zwiększeniem wartości. Nie wykorzystuj funkcji wbudowanych.

def epsilon_f16(f_16=np.float16): #zdefiniowanie funkcji liczącej epsilon dla skalaru np.float16
    epsilon = epsilon_last =  f_16(1) #zdefiniowanie zmiennych epsilon i epsilon_last
    #print(epsilon)
    while f_16(1)+f_16(epsilon) != f_16(1): #pętla, która powtarza się do momentu gdy epsilon dodany do jedności da jednośc
        epsilon_last = epsilon #przypisanie epislon_last wartości episilon z poprzedniej iteracji
        epsilon = f_16(epsilon) / f_16(2) #zmniejszenie wartści skalaru dla epsilona o skalar 2
    return epsilon_last #zwrócenie epsilona dla którego skończyła sie pętla 

def epsilon_f32(f_32=np.float32): #zdefiniowanie funkcji liczącej epsilon dla skalaru np.float32
    epsilon = epsilon_last =  f_32(1)
    while f_32(1)+f_32(epsilon) != f_32(1):
        epsilon_last = epsilon
        epsilon = f_32(epsilon) / f_32(2)
    return epsilon_last

def epsilon_f64(f_64=np.float64): #zdefiniowanie funkcji liczącej epsilon dla skalaru np.float64
    epsilon = epsilon_last =  f_64(1)
    while f_64(1)+f_64(epsilon) != f_64(1):
        epsilon_last = epsilon
        epsilon = f_64(epsilon) / f_64(2)
    return epsilon_last

print("Epsilon for float16:", epsilon_f16())
print("Epsilon for float32:",epsilon_f32())
print("Epsilon for float64:",type(epsilon_f64()))
```

```python
# ZADANIE 4

#Wygeneruj wartości funkcji logarytmu naturalnego w zakresie zmiennej od 0 do 5 dla 200 równomiernie, 
#monotoniczne rozłożonych wartości. Dodaj do niej szum normalny o wartości średniej równej 0 
#i odchyleniu standardowym równym 0.2. Narysuj wykres wygenerowanych wartości. 
#Następnie spróbuj usunąć szum poprzez uśrednienie wartości sąsiadujących stosując różny rozmiar okna (3, 7, 13). 
#Zignoruj wartości brzegowe. Narysuj wynik. Do wykresu dodaj siatkę, podpisy osi, legendę. 
#Jakie są wady takiego podejścia do usuwania szumu?


import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 200) #zdefiniowanie zbioru 200 równomiernie rozłozonych wartości od 0 do 5
y = np.log(x[1:]) # zdefiniowanie funkcji y jako logarytmu naturalnego
noise = np.random.normal(0, 0.2, 199) #wygenerowanie szumu normalnego, o średniej = 0 i odchyleniu = 0.2
y_noise = y + noise #zdefiniowanie funkcji y z uwzględnieniem szumu

def func_mean(N): #zdefiniowanie funkcji do liczenia średniej sąsiadujących wartości funkcji
    # N - rozmiar okna np. 3, 7, 13
    y_mean = np.array([]) #zdefiniowanie pustej tablicy
    for i in range(1,199): #pętla zapełniająca licząca wartości średnie sąsiadów i zapelniajaca nimi tablice
        mean = np.mean(y_noise[i:i+N+1]) #obliczenie średniej
        y_mean = np.append(y_mean, mean) #dodanie do tablicy wartości wyliczonej średniej 
    return y_mean #zwrócenie tablicy z średnimi wartościami sąsiadów

    
plt.figure(figsize=(18,6)) #przedstawienie wykresów uśrednionych wartości funkcji z szumem

plt.subplot(1,3,1)
plt.plot(x[1:],y ,"g", label="original")
plt.plot(x[1:], y_noise,"b", label="+Noise", alpha=0.4)
plt.plot(x[1:199], func_mean(3),"m", label="Averaged N=3")
plt.title("Rozmiar okna: 3")
plt.legend(loc='lower right')
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.subplot(1,3,2)
plt.plot(x[1:],y,"g", label="original")
plt.plot(x[1:], y_noise,"b", label="+Noise", alpha=0.4)
plt.plot(x[1:199], func_mean(7),"m", label="Averaged N=7")
plt.title("Rozmiar okna: 7")
plt.legend(loc='lower right')
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.subplot(1,3,3)
plt.plot(x[1:],y ,"g", label="original")
plt.plot(x[1:], y_noise,"b", label="+Noise", alpha=0.4)
plt.plot(x[1:199], func_mean(13),"m", label="Averaged N=13")
plt.title("Rozmiar okna: 13")
plt.legend(loc='lower right')
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.show()


#Otrzymane wykresy pokazują, że dla wiekszego okna funkcja uśrednionych wartości przyjmuje blizszy kształt 
#do funkcji bez szumu
#Wada takiego usuwania szumu jest niedokładne przyblizenie wartości, widac to na ostatnim wykresie gdzie 
#krzywa średnich jest wyżej niż ta bez szumu, dla małych wartości x jest ona mniej dokładna 
#Logarytm naturalny dla 0 nie istnieje
#Wada - szum pozostaje nawet po zwiekszeniu rozmiaru okna, mrf jako lepsza metoda (markov random fields)
```

```python
#ZADANIE 5
#Powtórz Zadanie 3 wykorzystując bibliotekę PyTorch.

def epsilon_f16(f_16 = tc.HalfTensor([1])): #zdefiniowanie funkcji liczącej epsilon dla tensora tc.HalfTensor
    epsilon = epsilon_last =  f_16[0] #zdefiniowanie zmiennych epsilon i epsilon_last
    while f_16[0] + tc.HalfTensor([epsilon]) != f_16[0]: #pętla, która powtarza się do momentu gdy epsilon dodany do jedności da jednośc
        epsilon_last = epsilon #przypisanie epislon_last wartości episilon z poprzedniej iteracji
        epsilon = epsilon / (f_16[0] +1) #zmniejszenie wartści skalaru dla epsilona o tensor 2
    return epsilon_last #zwrócenie epsilona dla którego skończyła sie pętla 

def epsilon_f32(f_32 = tc.FloatTensor([1])): #zdefiniowanie funkcji liczącej epsilon dla tensora tc.FloatTensor
    epsilon = epsilon_last =  f_32[0]
    while f_32[0] + tc.FloatTensor([epsilon]) != f_32[0]:
        epsilon_last = epsilon
        epsilon = epsilon / (f_32[0] +1)
    return epsilon_last #zwrócenie epsilona dla którego skończyła sie pętla 

def epsilon_f64(f_64 = tc.DoubleTensor([1])): #zdefiniowanie funkcji liczącej epsilon dla tensora tc.DoubleTensor
    epsilon = epsilon_last =  f_64[0]
    while f_64[0] + tc.DoubleTensor([epsilon]) != f_64[0]:
        epsilon_last = epsilon
        epsilon = epsilon / (f_64[0] +1)
    return epsilon_last #zwrócenie epsilona dla którego skończyła sie pętla 


print("Epsilon for float16:", epsilon_f16().item())
print("Epsilon for float32:", epsilon_f32().item())
print("Epsilon for float64:", epsilon_f64().item())


```
