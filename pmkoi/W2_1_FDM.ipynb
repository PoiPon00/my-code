# Różniczkowanie numeryczne

Przeanalizujmy właściwości metody różnic skończonych. Na początek import potrzebnych bibliotek

```python
import numpy as np
import matplotlib.pyplot as plt
```

Na początku, policzmy pochodną zgodnie ze wzorem  $ f'(x)=\frac{f(x_{i})-f(x_{i-1})}{h} $. 
Przyjmijmy, że $ h=1, x_{1}=1, f(x_{1})=0.84147, x_{2}=2, f(x_{2})=0.90930 $ i wyliczmy $f'(x_{2})$

```python
h=1
x1=1
x2=2
f1=0.84147
f2=0.90930
df=(f2-f1)/h
display(df)
```

Często pochodną wyliczamy nie dla jednego punktu, ale np. dla całego sygnału. W związku z tym, wygodniej będzie przygotować funkcję, która będzie wyliczała pochodną. Żeby nie było za łatwo oraz żebyście mieli co robić na laboratorium funkcja będzie pozwalać na wyliczenie pierwszej pochodnej metodą różnic skończonych wstecz.

```python
def poch(f,a,h=0.01):
    '''
    Parameters
    ----------
    f : funkcja jednej zmiennej
    a : argumenty, dla któych wyznaczamy pochodną
    h : krok        
    '''
    return (f(a + h) - f(a))/h  
```

Wygenerujmy wektor x 

```python
x = np.linspace(0,5*np.pi,100)
```

Policzmy pochodną funkcji sin z wykorzystaniem naszej funkcji

```python
df = poch(np.sin,x,h=0.1571)
```

Wiemy, że pochodną funkcji $sin(x)$ jes $cos(x)$

```python
df_true = np.cos(x)

plt.figure()
plt.plot(x,df,'r.',label='Metoda różnic wstecz')
plt.plot(x,df_true,'b',label='Wartość prawdziwa')
plt.legend(loc='best')
plt.show()
```

A teraz zwiększmy h

```python
x2 = np.linspace(0,5*np.pi,20)
df2 = poch(np.sin,x2,h=0.7854)
df_true = np.cos(x)

plt.figure()
plt.plot(x,df,'r.',label='Metoda różnic wstecz h=0.1571')
plt.plot(x2,df2,'go',label='Metoda różnic wstecz h=0.7854')
plt.plot(x,df_true,'b',label='Wartość prawdziwa')
plt.legend(loc='best')
#plt.axis([2,4,-1,-0.75])
plt.show()
```

Zdefinujmy inną funkcję $1-e^{-x/0.5}$

```python
f = lambda x: (1-np.exp(-x/0.5))
x3 = np.linspace(0,5,1000)
df3 = poch(f,x3,h=0.5/1000)

plt.figure()
plt.plot(x3,f(x3),'r.',label='funkcja')
plt.plot(x3,df3,'g.',label='Metoda różnic wstecz')
plt.legend(loc='best')
#plt.axis([2,4,-1,-0.75])
plt.show()
```
